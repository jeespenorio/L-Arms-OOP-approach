```python
import pandas as pd

class Arms:
    def __init__(self, make, model, year, mpn, sku, position, product_type, vendor, vendor_code, vendor_notes):
        self.make = make
        self.model = model
        self.year = year
        self.mpn = mpn
        self.sku = sku
        self.position = position
        self.product_type = product_type
        self.vendor = vendor
        self.vendor_notes = vendor_notes
        self.vendor_code = vendor_code
        
    def to_dict(self):
        return {
            'Make': [self.make],
            'Model': [self.model],
            'Vendor': [self.vendor],
            'Vendor_Code': [self.vendor_code],
            'Product_Type': [self.product_type],
            'Vendor_Notes': [self.vendor_notes],
            'Year': [self.year],
            'MPN': [self.mpn],
            'SKU': [self.sku],
            'Position': [self.position]
        }
```

```python
class ArmsGenerator:
    def __init__(self, arms_template, list_models, years):
        self.arms_template = arms_template
        self.list_models = list_models
        self.years = years
        
    def generate_arms(self):
        data = []
        for model in self.list_models:
            for year in self.years:
                arms_data = {
                    'Make': self.arms_template.make,
                    'Model': model,
                    'Vendor': self.arms_template.vendor,
                    'Vendor_Notes': self.arms_template.vendor_notes,
                    'Vendor_Code': self.arms_template.vendor_code,
                    'MPN': self.arms_template.mpn,
                    'SKU': self.arms_template.sku,
                    'Position': self.arms_template.position,
                    'Product_Type': self.arms_template.product_type,
                    'Year': year
                }
                data.append(arms_data)
        return pd.DataFrame(data)
    
```
